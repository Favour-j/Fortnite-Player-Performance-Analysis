```python
import pandas as pd;
import numpy as np; 
import seaborn as sns; 
import matplotlib.pyplot as plt
```

```python
# load the data
fortnite = pd.read_csv(r'C:\Users\mrsid\Documents\fortnite_100.csv')
```

```python
print(fortnite.head(10))
print(fortnite.info())
print(fortnite.describe())
```

```python
#check for nulls 
print(fortnite.isnull().sum())

#check for unique values in categorical columns
print(fortnite['Platform'].unique())
print(fortnite['Retention_Day_7'].unique())
```

```python
#find duplicate rows
duplicates = fortnite.duplicated()
print(f"Number of duplicate rows: {duplicates.sum()}")
```

```python
#drop duplicate rows
fortnite = fortnite.drop_duplicates()
```

```python
#standardize categorical data (Case string)
fortnite['Platform'] = fortnite['Platform'].str.upper().str.strip()
fortnite['platform'] = fortnite['Platform'].replace({'XBOX ONE': 'XBOX', 'PS5': 'PLAYSTATION', 'pc': 'PC'})
fortnite['Region'] = fortnite['Region'].replace({'N/A' : np.nan})
```

```python
#fix boolean values in Retention_Day_7 and Retention_Day_30
boolean_map = {'TRUE': True, 'T': True, '"TRUE"': True, 'FALSE': False, 'False': False, 'F': False, 'NA': False}
fortnite['Retention_Day_7'] = fortnite['Retention_Day_7'].map(boolean_map).fillna(False)
fortnite['Retention_Day_30'] = fortnite['Retention_Day_30'].map(boolean_map).fillna(False)
```

```python
# convert numeric types
fortnite['Avg_Session_Duration_Min'] = pd.to_numeric(fortnite['Avg_Session_Duration_Min'], errors='coerce')
fortnite['Lifetime_Wins'] = pd.to_numeric(fortnite['Lifetime_Wins'], errors='coerce')
# fill missing wins with median to avoid skewing the data with outliers
fortnite['Lifetime_Wins'] = fortnite['Lifetime_Wins'].fillna(fortnite['Lifetime_Wins'].median())
```

```python
#creating a win rate metric
fortnite['Win_Rate'] = ((fortnite['Lifetime_Wins'] / fortnite['Total_Matches']) * 100).round(2)
```

```python
# save cleaned data to new csv
fortnite.to_csv(r'C:\Users\mrsid\Documents\fortnite_100_cleaned.csv', index=False)
```
